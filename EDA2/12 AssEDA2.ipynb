### 1. Data Preprocessing 

```python
# Replace ' ?' with NaN and then impute with mode
df.replace(' ?', np.nan, inplace=True)
for col in ['workclass', 'occupation', 'native_country']:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

**Handling Missing Values**
Missing values were imputed using the Mode (most frequent value) for each respective categorical column. This ensures that the dataset remains complete while preserving the original distribution as much as possible.

```python
# Numerical columns
num_cols = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']
```

```python
# Standard Scaling
scaler_std = StandardScaler()
df_std = df.copy()
df_std[num_cols] = scaler_std.fit_transform(df[num_cols])
```

```python
# Min-Max Scaling
scaler_minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[num_cols] = scaler_minmax.fit_transform(df[num_cols])
```

**Feature Scaling**
Two primary scaling techniques were applied to numerical features:

**Standard Scaling ($Z$-score normalization):** Centers the data (mean = 0) and scales to unit variance (std = 1).\
**Min-Max Scaling:** Rescales the data to a fixed range, typically $[0, 1]$.

Scenarios for Each Technique:

**Standard Scaling:** Preferred for algorithms that assume normally distributed data (e.g., Linear Regression, Logistic Regression) and is more robust to outliers compared to Min-Max.

**Min-Max Scaling:** Preferred for algorithms that do not assume any distribution and require features to be on the same scale (e.g., K-Nearest Neighbors, Neural Networks). It is also essential when data must be bounded (e.g., image processing).

### 2. Encoding

**Encoding Techniques**\
Categorical variables were encoded based on their cardinality (number of unique categories):\
**One-Hot Encoding (OHE):** Applied to sex and income ($< 5$ categories).\
**Label Encoding (LE):** Applied to workclass, education, marital_status, occupation, relationship, race, and native_country ($\ge 5$ categories).

```python
cat_cols = df.select_dtypes(include=['object']).columns
unique_counts = {col: df[col].nunique() for col in cat_cols}
print("Unique counts for categorical columns:", unique_counts)
```

```python
# One-Hot Encoding for < 5 categories
# Label Encoding for >= 5 categories
df_encoded = df.copy()
label_encoder = LabelEncoder()
```

```python
ohe_cols = [col for col in cat_cols if df[col].nunique() < 5]
le_cols = [col for col in cat_cols if df[col].nunique() >= 5]

print("Columns for OHE (<5):", ohe_cols)
print("Columns for LE (>=5):", le_cols)
```

```python
# Apply One-Hot
df_encoded = pd.get_dummies(df_encoded, columns=ohe_cols)
```

```python
# Apply Label Encoding
for col in le_cols:
    df_encoded[col] = label_encoder.fit_transform(df_encoded[col])
```

### 3. Feature Engineering

**Feature Engineering**\
To enhance model predictive power, the following transformations and features were created:\
**1)capital_diff ($Capital\ Gain - Capital\ Loss$):**\
Rationale: Instead of looking at gains and losses separately, this captures the net investment success of an individual, which is a stronger indicator of wealth.\
**2)age_edu_interaction ($Age \times Education\ Num$):**\
Rationale: The impact of education on income often compounds with age. A highly educated older individual is statistically more likely to be in a high-income bracket than a highly educated younger individual.\
**3)Log Transformation (capital_gain_log):**\
Choice: Applied to capital_gain because it is highly right-skewed with many zeros and extreme outliers.\
Justification: Transformation using $\log(1+x)$ compresses the range of extreme values, making the distribution more Gaussian-like and helping linear models perform better.

```python
# 1. Net Capital
df_encoded['capital_diff'] = df_encoded['capital_gain'] - df_encoded['capital_loss']
```

```python
# 2. Education-Age Interaction or Binary Gender
df_encoded['age_edu_interaction'] = df_encoded['age'] * df_encoded['education_num']
```

```python
# Log transformation for skewed feature (capital_gain)
# Log(1+x) to handle zeros
df_encoded['capital_gain_log'] = np.log1p(df_encoded['capital_gain'])
```

### 4. Feature Selection

**Feature Selection and Outlier Detection**\
**Isolation Forest for Outliers**\
The Isolation Forest algorithm was used to identify anomalies. Approximately 1,628 outliers ($5\%$) were detected and removed.\
**Impact of Outliers:** Outliers can disproportionately influence the mean and variance, leading to biased scaling. In modeling, they can pull the decision boundary towards noise, resulting in poor generalization and overfitting.

```python
# Isolation Forest for outliers
iso = IsolationForest(contamination=0.05, random_state=42)
outliers = iso.fit_predict(df_encoded.select_dtypes(include=[np.number]))
df_no_outliers = df_encoded[outliers == 1]

print(f"Removed {np.sum(outliers == -1)} outliers.")
```

Since the ppscore library is a specialized tool, Mutual Information (MI) was used as a proxy to assess relationships.\
**Correlation Matrix (Pearson):** Measures linear relationships. It showed a strong correlation between education_num and income ($0.335$). However, it fails to capture non-linear dependencies.\
**PPS / Mutual Information:** MI captures any type of statistical dependency (linear or non-linear).\
**Findings:** MI revealed that marital_status and relationship have higher predictive power than age or hours_per_week, despite age having a higher linear correlation score.

```python
# Correlation Matrix
corr_matrix = df_encoded.corr()
corr_matrix
```

Predictive Power Score (PPS) vs. Correlation Matrix\
**Comparison:** The correlation matrix is excellent for identifying simple linear trends, whereas PPS/MI is superior for identifying complex patterns that modern non-linear models (like Random Forests or XGBoost) can exploit.

### Compare its findings with the correlation matrix
